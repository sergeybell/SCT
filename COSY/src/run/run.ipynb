```python
fox_files = ["cosy.fox", "utilities.fox", "elements.fox", "header.fox"]
exe_name = "cosy.exe"
bat_name = "pre_run_cosy.bat"

with open(bat_name, "w", encoding="cp866") as f:
    f.write("@echo off\n")
    for fox in fox_files:
        f.write(f"echo ========================================\n")
        f.write(f"echo RUNNING: {fox}\n")
        f.write(f"echo ========================================\n")
        # Переход в корень, запуск и возврат в run
        f.write(f"cd ..\n")
        f.write(f"{exe_name} {fox}\n")
        f.write(f"cd run\n")
        f.write(f"echo.\n") # Пустая строка для читаемости

print(f"✅ Файл {bat_name} обновлен и готов к работе.")
```
